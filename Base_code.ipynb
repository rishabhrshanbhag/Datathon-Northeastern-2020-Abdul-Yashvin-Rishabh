```python
import numpy as np
import pandas as pd
```

```python
#importing New York data.
covid_data = pd.read_csv('../covid-new-york.csv')
#importing Google Mobility data
mobility = pd.read_csv('../us-mobility.csv')
```

```python
#Selecting data from New York
mobility = mobility[mobility.state == 'New York']
#Converting data format in mobility to match covid data ex: 'Albany county' -> 'Albany'
mobility['county'] = mobility.county.apply(lambda x: str(x).split()[0])
#Dropping unwanted columns from the dataframes
mobility = mobility.drop(['Unnamed: 0','state'], axis=1)
covid_data = covid_data.drop(['Unnamed: 0'],axis=1)
```

```python
#Merge Covid data with mobility data on date and county.
df = pd.merge(covid_data, mobility,  how='left', left_on=['test_date','county'], right_on = ['date','county',])
#drop all rows with no data in mobility dataframe
df = df[df['date'].notna()]
#drop test_date as it is the same as the date column
df = df.drop(['test_date'],axis=1)
#save this data for future use.
df.to_csv('../data.csv')
```
