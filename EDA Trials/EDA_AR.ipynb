```python
import plotly.express as px
import scipy
from scipy.signal import savgol_filter

df_no_na = df[df['parks'].notna()]
df_no_na = df_no_na.groupby('date')['parks'].sum().to_frame().reset_index()

fig = px.line(x=df_no_na['date'],
              y=savgol_filter(df_no_na['parks'],33,3),
              title='Mobility of Parks')
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

df_no_na = df[df['parks'].notna()]
df_mob_date = df_no_na.groupby('date')['parks'].sum().to_frame().reset_index()
df_case_date = df_no_na.groupby('date')['new_positives'].sum().to_frame().reset_index()
df_no_na = pd.merge(df_mob_date, df_case_date,  how='left', left_on=['date'], right_on = ['date'])

fig = make_subplots(rows=2, cols=1)

fig.add_trace(
    go.Scatter(x=df_no_na['date'], y=savgol_filter(df_no_na['parks'],33,3)),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=df_no_na['date'], y=savgol_filter(df_no_na['new_positives'],33,3)),
    row=2, col=1
)

fig.update_layout(height=800, width=800, title_text='Mobility in Parks Compared with number of cases')
fig.show()
```
